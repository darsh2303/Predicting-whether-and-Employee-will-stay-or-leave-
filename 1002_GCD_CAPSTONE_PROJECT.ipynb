# Project description

**HR department at a software company is looking to try a new initiative to retain employees by using data to predict whether an employee is likely to leave as the current practice of taking exit interviews do provide insights, but its too late to work on it. Also these insights can't be aggregated and interlaced across all employees who have left.
Trying the new initiative once these employees are identified, HR can be more proactive in reaching out to them before it's too late.**

# Project Goals

**Given datasets of past employees and their status our task is to build a classification model using the datasets to Predict whether an employee will stay or leave.**

# Data Description

**The Business Intelligence Analysts of the Company provided you three datasets that contain information about past employees and their status (still employed or already left).**

- **department_data**

    This dataset contains information about each department. The schema of the dataset is as follows:

    dept_id : Unique Department Code<br>
    dept_name : Name of the Department<br>
    dept_head : Name of the Head of the Department<br>


- **employee_details_data**

    
    This dataset consists of Employee ID, their Age, Gender and Marital Status. The schema of this dataset is as follows:

     employee_id : Unique ID Number for each employee<br>
     age : Age of the employee<br>
     gender : Gender of the employee<br>
     marital_status : Marital Status of the employee<br>


- **employee_data**

    This dataset consists of each employee’s Administrative Information, Workload Information, Mutual Evaluation Information and Status.


- **Target variable**

    status – Current employment status (Employed / Left)

- **Administrative information**

    department – Department to which the employees belong(ed) to<br>
    salary – Salary level with respect to rest of their department<br>
    tenure – Number of years at the company<br>
    recently_promoted – Was the employee promoted in the last 3 years?<br>
    employee_id – Unique ID Number for each employee<br>


- **Workload information**

    n_projects – Number of projects employee has worked on<br>
    avg_monthly_hrs – Average number of hours worked per month<br>

- **Mutual evaluation information**

    satisfaction – Score for employee’s satisfaction with the company (higher is better)<br>
    last_evaluation – Score for most recent evaluation of employee (higher is better)<br>
    filed_complaint – Has the employee filed a formal complaint in the last 3 years?

## Treating Missing values in Employee Data.

**Creating copies of all the datasets.**

```python
department_data_copy = department_data.copy(deep=True)
employee_details_data_copy = employee_details_data.copy(deep=True)
employee_data_copy = employee_data.copy(deep=True)
```

```python
#Rename Column name
employee_data_copy.rename(columns={'department':'dept_id'}, inplace=True)
```

```python
#creating new department "D00-UNK"
employee_data_copy['dept_id'].fillna('D00-UNK' , inplace=True) 
```

```python
#droping duplicate rows

Nrows_bfr =len(employee_data_copy)
print('No. of Rows before droping duplicate= ',Nrows_bfr)

employee_data_copy.drop_duplicates(inplace=True)

Nrows_afr = len(employee_data_copy)
print('No. of Rows before droping duplicate= ',Nrows_afr)

print('No. of rows dropped = ', Nrows_bfr-Nrows_afr)
```

```python
employee_data_copy.isnull().sum()
```

```python
#replacing null values in filed_complaint & recently_promoted with 0 and convert them to categorical data.

employee_data_copy['filed_complaint'] = employee_data_copy['filed_complaint'].replace(np.nan,0)
employee_data_copy['recently_promoted'] = employee_data_copy['recently_promoted'].replace(np.nan,0)
```

```python
employee_data_copy[['filed_complaint','recently_promoted']].astype('category')
```

- filed complaint & promoted column null values are replaced by 0 and converted to categorical data with 1 representing a promotion or filed complaint and 0 reprsenting no promotion and no complaint.

```python
employee_data_copy.info()
```

```python
#Replacing null values in satisfaction & Tenure columns with 0.

employee_data_copy['satisfaction'] = employee_data_copy['satisfaction'].replace(np.nan,0)
employee_data_copy['tenure'] = employee_data_copy['tenure'].replace(np.nan,0)
```

```python
employee_data_copy.isnull().sum()
```

```python
#Replacing '-IT' with 'D00-IT'

employee_data_copy.dept_id.replace({'-IT':'D00-IT'}, inplace=True)
employee_data_copy.dept_id.unique()
```

```python
#replacing Salary values as Low = 0; Medium=1; High=2
sal_replace = {'low':0, 'medium':1, 'high':2}
employee_data_copy['salary'] = employee_data_copy.salary.map(sal_replace)
```

```python
employee_data_copy.salary.unique()
```

```python
#replacing Employee status as left = 0, employed = 1

status_replace = {'Left':1,'Employed':0}
employee_data_copy['status'] = employee_data_copy.status.map(status_replace)
```

```python
employee_data_copy.status.unique()
```

```python
#converting Department names into numeric

nos_list = list(range(0,12,1))
print(nos_list)
dept_list = employee_data_copy.dept_id.unique().tolist()
print(dept_list)
dept_dict = dict(zip(dept_list,nos_list))
print(dept_dict)

employee_data_copy['dept_id']=employee_data_copy.dept_id.map(dept_dict)
employee_data_copy.dept_id.unique()
```

**We use KNN Imputer to fill missing values for last_evaluation column.**

```python
from sklearn.impute import KNNImputer
employee_knn_imputed = employee_data_copy.copy(deep=True)
fig, ax = plt.subplots(figsize=(16, 8))
# Plot the original distribution
sns.kdeplot(employee_data_copy.last_evaluation, label="Original Distribution");

for k in range(3,10,2):
    knn_imp = KNNImputer(n_neighbors=k)
    employee_knn_imputed.loc[:, :] = knn_imp.fit_transform(employee_data_copy)
    sns.kdeplot(employee_knn_imputed.last_evaluation, label=f"Imputed Dist with k={k}");

plt.legend();
```

- **We tried KNNimputer with different values of k and k=3 looks closest to the original ditribution.**

```python
imputer = KNNImputer(n_neighbors=3)
imputed = imputer.fit_transform(employee_knn_imputed)
emp_data_df_imputed = pd.DataFrame(imputed, columns=employee_knn_imputed.columns)
```

```python
emp_data_df_imputed
```

```python
#getiing dept_id values back

dept_dict_new = dict([(value,key) for key,value in dept_dict.items()])
print(dept_dict_new)

emp_data_df_imputed['dept_id']=emp_data_df_imputed.dept_id.map(dept_dict_new)
emp_data_df_imputed.dept_id.unique()
```

```python
emp_data_df_imputed.isnull().sum()
```

- All missing values treated.
- For treating missing values in last evaluation column we used KNN imputer. Iterative imputer was also tried for the same, with no major difference in the ML models.

```python
def display_confusion_matrix(y_actual, y_preds):
    Confusion_Matrix = pd.DataFrame(confusion_matrix(y_actual, y_preds))
    
    Confusion_Matrix.index = ['Actual Employed', 'Actual Left']
    Confusion_Matrix.columns = ['Predicted Employed', 'Predicted Left']
    
    print('Confusion Matrix')
    print(Confusion_Matrix)
```

```python
def Treat_unseen(data_t):
    data_t.rename(columns={'department':'dept_id'}, inplace=True)
    data_t.dept_id.fillna('D00-UNK',inplace=True)
    data_t.dept_id.replace({'-IT':'D00-IT'},inplace=True)
    data_t.filed_complaint.fillna(0,inplace=True)
    data_t.recently_promoted.fillna(0,inplace=True)
    data_t.satisfaction.fillna(0,inplace=True)
    data_t.tenure.fillna(0,inplace=True)
    #label encoding
    data_t['dept_id']=data_t.dept_id.map(dept_dict)
    data_t['salary'] = data_t.salary.map(sal_replace)
    
    data_t_imputed = data_t.copy(deep=True)
    #filling last_evaluation using KNN imputer
    imputer = KNNImputer(n_neighbors=3)
    imputed = imputer.fit_transform(data_t_imputed)
    data_t_imputed = pd.DataFrame(imputed, columns=data_t_imputed.columns)
    
    #Merging unseen data with emp details dataset
    data_t_final = pd.merge(data_t_imputed, employee_details_data_copy, on='employee_id', how='left')
    
    #label encoding
    data_t_final['gender'] = data_t_final.gender.map(gender_replace)
    data_t_final['marital_status'] = data_t_final.marital_status.map(marital_replace)
    
    return data_t_final
```
